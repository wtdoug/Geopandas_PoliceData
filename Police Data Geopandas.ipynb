```python
graffiti = gpd.read_file(r"C:\Users\wtdud\PycharmProjects\graffitiReportsGPD\PoliceData.gdb\PoliceData.gdb", layer = "GraffitiIncidents")
patrolZone = gpd.read_file(r"C:\Users\wtdud\PycharmProjects\graffitiReportsGPD\PoliceData.gdb\PoliceData.gdb", layer = "PatrolZones")
```

```python
patrolZones_with_graffiti = gpd.sjoin(graffiti, patrolZone, how='inner', predicate='within')
```

```python
patrolZones_with_graffiti.NAME.value_counts()
```

```python
patrolZones_with_graffiti
```

```python
df = patrolZones_with_graffiti.drop(["COMMENTS", "geometry", "index_right", "POPULATION","SHAPE_Length"], axis=1)
df
```

```python
df["sqMile_area"] = df["SHAPE_Area"]/2589988.11
df.drop(["SHAPE_Area"], axis=1)
```

```python
SCfilt = (df['NAME'] == "South Central")
df.loc[SCfilt, 'INCIDENTS'] = '23'
NCfilt = (df['NAME'] == "North Central")
df.loc[NCfilt, 'INCIDENTS'] = '21'
SEfilt = (df['NAME'] == "Southeast")
df.loc[SEfilt, 'INCIDENTS'] = '21'
ESidefilt = (df['NAME'] == "Eastside")
df.loc[ESidefilt, 'INCIDENTS'] = '20'
NWfilt = (df['NAME'] == "Northwest")
df.loc[NWfilt, 'INCIDENTS'] = '13'
SWfilt = (df['NAME'] == "Southwest")
df.loc[SWfilt, 'INCIDENTS'] = '12'
NEfilt = (df['NAME'] == "Northeast")
df.loc[NEfilt, 'INCIDENTS'] = '10'
DTfilt = (df['NAME'] == "Downtown")
df.loc[DTfilt, 'INCIDENTS'] = '10'
Cfilt = (df['NAME'] == "Central")
df.loc[Cfilt, 'INCIDENTS'] = '6'
```

```python
df
```

```python
group = df.groupby(['NAME'])
df['PRIORITY'] = group['INCIDENTS'].transform('count')/df['sqMile_area']
```
